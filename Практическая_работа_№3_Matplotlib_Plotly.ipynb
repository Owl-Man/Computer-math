3. Используя библиотеку plotly express, создайте интерактивный график линий с помощью функции [`plotly.express.line`](https://plotly.com/python-api-reference/generated/plotly.express.line.html).
4. В качестве параметров функции `line` передайте массив месяцев в качестве оси `x`, массивы температур в качестве оси `y`, а также названия городов в качестве параметра `color`, чтобы разделить линии по группам .

```python
import numpy as np
import plotly.express as px
import pandas as pd

months = np.arange(1, 13)

moscow_temps = np.random.uniform(low=-10, high=30, size=12)
paris_temps = np.random.uniform(low=-10, high=30, size=12)
newyork_temps = np.random.uniform(low=-10, high=30, size=12)

data = {
    'Month': np.tile(months, 3),
    'Temperature': np.concatenate([moscow_temps, paris_temps, newyork_temps]),
    'City': ['Moscow']*12 + ['Paris']*12 + ['New York']*12
}
df = pd.DataFrame(data)

fig = px.line(df, x='Month', y='Temperature', color='City', title='Средняя температура в городах по месяцам')

fig.update_layout(
    xaxis_title='Месяц',
    yaxis_title='Температура (°C)',
    legend_title='Город',
    title_x=0.5
)

fig.show()
```

2. Создайте массив `y`, который является логарифмом всех значений `x`.
3. На первом подграфике постройте график `y` в зависимости от `x` с использованием линейной шкалы осей.
4. На втором подграфике постройте тот же график, но с логарифмической шкалой по оси $x$ ([`.xscale()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xscale.html)).
5. Сравните два графика и сделайте выводы о различиях между шкалами.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Создаем массив x из 100 точек от 1 до 1000
x = np.linspace(1, 1000, 100)

# 2. Создаем массив y как логарифм значений x (натуральный логарифм)
y = np.log(x)

# Создаем фигуру с двумя подграфиками, расположенными вертикально
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# 3. Первый подграфик: линейная шкала
ax1.plot(x, y, color='blue')
ax1.set_title('График y = log(x) с линейной шкалой')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# 4. Второй подграфик: логарифмическая шкала по оси X
ax2.plot(x, y, color='green')
ax2.set_xscale('log')  # Устанавливаем логарифмическую шкалу для оси X
ax2.set_title('График y = log(x) с логарифмической шкалой по оси X')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# Увеличиваем расстояние между подграфиками
plt.subplots_adjust(hspace=0.4)

# Добавляем общий заголовок
fig.suptitle('Сравнение линейной и логарифмической шкалы', fontsize=16)

# Отображаем график
plt.show()
```

### Задание 2.6: Создание 3D графиков

1. Используя NumPy, создайте двумерные массивы `X` и `Y` с помощью функции `numpy.meshgrid()`. Задайте интересующий диапазон значений для каждой оси.
2. На основе массивов `X` и `Y` сгенерируйте массив `Z`, используя любое математическое выражение (например, `Z = numpy.sin(X) + numpy.cos(Y)`).
3. Используя Matplotlib, создайте фигуру, а затем объект для трехмерного графика с помощью [`fig.add_subplot(111, projection='3d')`](https://matplotlib.org/3.2.1/tutorials/toolkits/mplot3d.html)`.
4. Отобразите сгенерированные данные в виде поверхности, используя метод `plot_surface`.
5. Настройте визуальные аспекты графика - прозрачность `set_alpha()` и угол обзора `view_init()`.
6. Добавьте подписи к осям и заголовок графика.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y) 

Z = np.sin(X) + np.cos(Y) 

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis')

surf.set_alpha(0.8)
ax.view_init(elev=30, azim=45)

fig.colorbar(surf, ax=ax, label='Z значение')

ax.set_xlabel('X ось')
ax.set_ylabel('Y ось')
ax.set_zlabel('Z ось')
ax.set_title('3D график поверхности Z = sin(X) + cos(Y)')

plt.show()
```

### Задание 2.7: Трехмерный график рассеяния

1. Используя библиотеку numpy, сгенерируйте три одномерных массива, представляющих координаты `X`, `Y` и `Z` случайных точек в пространстве. Пусть количество точек равно 100, а диапазон значений каждой координаты от 0 до 10.
2. Используя библиотеку plotly express, создайте трехмерный график рассеяния с помощью функции `px.scatter_3d`.
3. В качестве параметров функции `scatter_3d` передайте массивы `X`, `Y` и `Z`.
4. Настройте внешний вид графика, изменяя цвет и размер точек `fig.update_traces()`, а также добавляя подписи к осям и заголовок графика `fig.update_layout()`.






```python
import numpy as np
import plotly.express as px
import pandas as pd

x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
z = np.random.uniform(0, 10, 100)  

data = {
    'X': x,
    'Y': y,
    'Z': z
}
df = pd.DataFrame(data)

fig = px.scatter_3d(df, 
                    x='X', 
                    y='Y', 
                    z='Z',
                    title='Трёхмерный график рассеяния случайных точек')

fig.update_traces(marker=dict(size=5,
                              color='purple',
                              opacity=0.8))

fig.update_layout(
    scene=dict(
        xaxis_title='X ось',
        yaxis_title='Y ось',
        zaxis_title='Z ось'
    ),
    title_x=0.5
)

fig.show()
```

#### Описание модели SIR:

Модель SIR (Susceptible, Infected, Recovered) представляет собой классическую эпидемиологическую модель, которая разделяет популяцию на три основные группы:
- **S (Susceptible)**: Часть популяции, восприимчивая к инфекции, но еще не зараженная.
- **I (Infected)**: Часть популяции, зараженная инфекцией и способная передавать ее другим.
- **R (Recovered)**: Часть популяции, выздоровевшая после инфекции и получившая иммунитет.

#### Математический аппарат модели SIR:

1. **Дифференциальные уравнения модели SIR:**

   - $\frac{dS}{dt} = -\beta \frac{SI}{N}$ — уменьшение числа восприимчивых из-за заражения.
   - $\frac{dI}{dt} = \beta \frac{SI}{N} - \gamma I$ — изменение числа инфицированных, учитывая новые случаи заражения и выздоровление.
   - $\frac{dR}{dt} = \gamma I$ — увеличение числа выздоровевших благодаря выздоровлению инфицированных.

   Где:
   - $N$ — общая численность популяции (постоянная величина).
   - $\beta$ — коэффициент заразности, определяющий вероятность передачи инфекции от инфицированных к восприимчивым.
   - $\gamma$ — коэффициент выздоровления, обратный среднему периоду инфекционности.

  - `gamma`: Скорость выздоровления, указывает на частоту, с которой инфицированные особи переходят в состояние выздоровевших.
  - `S`: Количество восприимчивых к болезни в начале моделирования.

  **Методы:**
  - `__init__`: Конструктор класса, инициализирует все параметры модели и задает начальные условия.
  - `run`: Основной метод, выполняющий моделирование. Использует дифференциальные уравнения для расчета изменений в количестве восприимчивых, инфицированных и выздоровевших на каждом шаге времени.
  - `plot`: Метод для визуализации результатов моделирования. Создает интерактивный график, отображающий динамику изменения популяций во времени.

  **Характеристика методов:**
  - **Метод `run`** использует численные методы для решения системы дифференциальных уравнений, что позволяет проследить за изменениями в количестве восприимчивых, инфицированных и выздоровевших в зависимости от времени.

    В методе `run` класса `SIRModel`, используются следующие формулы для вычисления изменений количества восприимчивых (S), инфицированных (I) и выздоровевших (R) в популяции:

  - Изменение восприимчивых:
    $$\Delta S = -\beta \frac{SI}{N} \Delta t$$

  - Изменение инфицированных:
    $$\Delta I = \left(\beta \frac{SI}{N} - \gamma I\right) \Delta t$$

  - Изменение выздоровевших:
    $$\Delta R = \gamma I \Delta t$$

    Где:
    - $ \beta $ — коэффициент передачи инфекции, который определяет вероятность заражения.
    - $ \gamma $ — скорость выздоровления, которая указывает на частоту перехода инфицированных в состояние выздоровевших.
    - $ N $ — общая численность популяции.
    - $ \Delta t $ — шаг по времени.

    Эти формулы применяются на каждом шаге времени в цикле моделирования для обновления значений $ S $, $ I $, и $ R $.

  - **Метод `plot`** применяет библиотеку Plotly Express для создания графика, что обеспечивает наглядность и интерактивность представления данных. Пользователь может взаимодействовать с графиком, например, изменяя масштаб или просматривая точные значения популяций в различные моменты времени.

```python
# Импортируем необходимые библиотеки
import numpy as np
import plotly.express as px

# Создаем класс для моделирования распространения заболевания
class SIRModel:
    def __init__(self, N, beta, gamma, S0, I0, R0, t_max, dt):
        self.N = N  # Общая численность популяции
        self.beta = beta  # Коэффициент передачи инфекции
        self.gamma = gamma  # Скорость выздоровления
        self.S = S0  # Начальное число восприимчивых
        self.I = I0  # Начальное число инфицированных
        self.R = R0  # Начальное число выздоровевших
        self.t_max = t_max  # Максимальное время моделирования
        self.dt = dt  # Шаг по времени
        self.t = 0  # Текущее время
        
        # Списки для хранения истории
        self.S_list = [S0]
        self.I_list = [I0]
        self.R_list = [R0]
        self.t_list = [0]
        pass

    def run(self):
        # Запускаем моделирование
        while self.t < self.t_max:
            # Рассчитываем изменения популяций за один шаг по времени
            dS = -self.beta * self.S * self.I / self.N * self.dt
            dI = (self.beta * self.S * self.I / self.N - self.gamma * self.I) * self.dt
            dR = self.gamma * self.I * self.dt
            # Обновляем популяции с учетом изменений
            self.S += dS
            self.I += dI
            self.R += dR
            # Обновляем время
            self.t += self.dt
            # Добавляем результаты в списки
            self.S_list.append(self.S)
            self.I_list.append(self.I)
            self.R_list.append(self.R)
            self.t_list.append(self.t)


    def plot(self):
        print()
        # Преобразуем списки в массивы NumPy
        t_array = np.array(self.t_list)
        S_array = np.array(self.S_list)
        I_array = np.array(self.I_list)
        R_array = np.array(self.R_list)

        data = {
            'Time': np.tile(t_array, 3),
            'Population': np.concatenate([S_array, I_array, R_array]),
            'Group': ['Susceptible']*len(t_array) + ['Infected']*len(t_array) + ['Recovered']*len(t_array)
        }
        df = pd.DataFrame(data)
        
        # Добавляем подписи к легенде и осям
        fig = px.line(df, 
                      x='Time', 
                      y='Population', 
                      color='Group',
                      title='Динамика SIR модели')

        # Обновляем легенду с названиями групп
        fig.update_layout(
            xaxis_title='Время (дни)',
            yaxis_title='Численность популяции',
            legend_title='Группы',
            title_x=0.5 
        )

        # Показываем график
        fig.show()
```

4. **Особенности математического аппарата**

- **Класс `Prey` (Жертвы):**
  1. **Рост популяции жертв:** Прирост популяции жертв без учета хищников описывается уравнением $$ dx = \alpha x $$ где $ \alpha $ - коэффициент роста, $ x $ - текущее количество жертв.
  2. **Убыль популяции жертв:** Убыль популяции жертв из-за хищников описывается уравнением $$ dx = -\beta x y $$ где $ \beta $ - коэффициент смертности жертв от хищников, $ y $ - количество хищников.

- **Класс `Predator` (Хищники):**
  1. **Рост популяции хищников:** Прирост популяции хищников за счет питания жертвами описывается уравнением $$ dy = \delta x y $$ где $ \delta $ - коэффициент эффективности превращения пищи в новых хищников.
  2. **Убыль популяции хищников:** Убыль популяции хищников из-за естественной смертности описывается уравнением $$ dy = -\gamma y $$ где $ \gamma $ - коэффициент естественной смертности хищников.

- **Класс `EcoSystem` (Экосистема):**
  - В этом классе происходит моделирование взаимодействия между хищниками и жертвами. Используя метод `run`, модель выполняет итерации, на каждом шаге рассчитывая изменения в популяциях с помощью уравнений, описанных выше, и обновляя их значения.

```python
# Импортируем необходимые библиотеки
import numpy as np
import plotly.express as px
import matplotlib.pylab as pb

# Создаем класс для моделирования популяции жертв
class Prey:
    def __init__(self, x0, alpha):
        # Инициализируем начальное количество жертв и коэффициент роста
        self.x = x0  # Начальное количество жертв
        self.alpha = alpha  # Коэффициент рождаемости жертв

    def grow(self):
        # Рассчитываем прирост популяции жертв без учета хищников
        return self.alpha * self.x

    def die(self, y, beta):
        # Рассчитываем убыль популяции жертв из-за хищников
        return beta * self.x * y

    def update(self, dx):
        # Обновляем количество жертв с учетом прироста и убыли
        self.x += dx

# Создаем класс для моделирования популяции хищников
class Predator:
    def __init__(self, y0, gamma):
        # Инициализируем начальное количество хищников и коэффициент смертности
        self.y = y0  # Начальное количество хищников
        self.gamma = gamma  # Коэффициент смертности хищников

    def grow(self, x, delta):
        # Рассчитываем прирост популяции хищников за счет питания жертвами
        return delta * x * self.y

    def die(self):
        # Рассчитываем убыль популяции хищников из-за естественной смертности
        return self.gamma * self.y

    def update(self, dy):
        # Обновляем количество хищников с учетом прироста и убыли
        self.y += dy

# Создаем класс для моделирования динамики популяции хищник-жертва
class EcoSystem:
    def __init__(self, x0, y0, alpha, beta, gamma, delta, t_max, dt):
        # Инициализируем параметры модели
        self.prey = Prey(x0, alpha)  # Объект популяции жертв
        self.predator = Predator(y0, gamma)  # Объект популяции хищников
        self.beta = beta  # Коэффициент смертности жертв из-за хищников
        self.delta = delta  # Коэффициент роста хищников от жертв
        self.t_max = t_max  # Максимальное время моделирования
        self.dt = dt  # Шаг по времени
        self.t = 0  # Текущее время
        
        # Списки для хранения истории
        self.x_list = [x0]
        self.y_list = [y0]
        self.t_list = [0]

    def run(self):
        # Запускаем моделирование
        while self.t < self.t_max:
            # Рассчитываем изменения популяций за один шаг по времени
            x = self.prey.x
            y = self.predator.y
            dx_dt = self.prey.grow() - self.prey.die(y, self.beta)
            dx = dx_dt * self.dt
            dy_dt = self.predator.grow(x, self.delta) - self.predator.die()
            dy = dy_dt * self.dt
            # Обновляем популяции с учетом изменений
            self.prey.update(dx)
            self.predator.update(dy)
            # Обновляем время
            self.t += self.dt
            # Добавляем результаты в списки
            self.x_list.append(self.prey.x)
            self.y_list.append(self.predator.y)
            self.t_list.append(self.t)

    def plot(self):
      # Преобразуем списки в массивы NumPy
      t_array = np.array(self.t_list)
      x_array = np.array(self.x_list)
      y_array = np.array(self.y_list)
      # Создаем линейный график изменения популяций во времени
    #   df_time = pd.DataFrame({
    #         'Time': np.tile(t_array, 2),
    #         'Population': np.concatenate([x_array, y_array]),
    #         'Group': ['Prey']*len(t_array) + ['Predator']*len(t_array)})
      fig1 = px.line(x=np.tile(t_array, 2), 
                       y=np.concatenate([x_array, y_array]), 
                       color=['Prey']*len(t_array) + ['Predator']*len(t_array),
                       title='Динамика популяций хищник-жертва')
      fig1.update_layout(
            xaxis_title='Время',
            yaxis_title='Размер популяции',
            legend_title='Группы',
            title_x=0.5
        )
    #   # Создаем фазовый портрет
    #   df_phase = pd.DataFrame({
    #         'Prey': x_array,
    #         'Predator': y_array})
      fig2 = px.scatter(x=x_array, 
                          y=y_array,
                          title='Фазовый портрет системы хищник-жертва')
      fig2.update_layout(
            xaxis_title='Численность жертв',
            yaxis_title='Численность хищников',
            title_x=0.5
        )
      # Отображаем оба графика
      fig1.show()
      fig2.show()

```

```python
# Создаем объект экосистемы с заданными параметрами
ecosystem = EcoSystem(x0=10, y0=5, alpha=0.1, beta=0.01, gamma=0.1, delta=0.01, t_max=1000, dt=0.1)
# Запускаем моделирование
ecosystem.run()
# Визуализируем результаты
ecosystem.plot()
```
